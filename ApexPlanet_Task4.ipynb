```python
import pandas as pd

df = pd.read_csv('/content/Sample - Superstore.csv', encoding='latin1')
df.head()
```

```python
print("Rows:", df.shape[0])
print("Columns:", df.shape[1])

df.info()
```

```python
df.isnull().sum()
```

```python
print("Before:", df.shape)

df = df.drop_duplicates()

print("After:", df.shape)
```

```python
import matplotlib.pyplot as plt

df['Category'].value_counts().plot(kind='bar')

plt.title('Category Distribution')
plt.xlabel('Category')
plt.ylabel('Count')

plt.show()
```

```python
df.groupby('Category')['Sales'].sum().plot(kind='bar')

plt.title('Total Sales by Category')
plt.xlabel('Category')
plt.ylabel('Sales')

plt.show()
```

```python
technology = df[df['Category'] == 'Technology']['Sales']
furniture = df[df['Category'] == 'Furniture']['Sales']
```

```python
from scipy.stats import ttest_ind

t_stat, p_value = ttest_ind(technology, furniture)

print("T-statistic:", t_stat)
print("P-value:", p_value)
```

## Hypothesis Testing

**Null Hypothesis (H₀):** There is no significant difference in average sales between Technology and Furniture categories.

**Alternative Hypothesis (H₁):** There is a significant difference in average sales between Technology and Furniture categories.

**Test Used:** Independent Samples T-Test

**T-statistic:** 3.8431

**P-value:** 0.000123

### Conclusion

Since the p-value (0.000123) is less than 0.05, we reject the null hypothesis.

Therefore, there is a statistically significant difference in average sales between the Technology and Furniture categories.

**Business Conclusion:** The sales performance of Technology and Furniture categories differs significantly. The company can focus more on the higher-performing category to improve business revenue.
